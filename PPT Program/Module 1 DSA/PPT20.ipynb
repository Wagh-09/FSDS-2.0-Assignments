1. Given a binary tree, task is to find subtree with maximum sum in tree.

Examples:

Input : Â  Â  Â  1

/ Â  \

2Â  Â  Â  3

/ \Â  Â  / \

4 Â  5Â  6 Â  7

Output : 28

As all the tree elements are positive,

the largest subtree sum is equal to

sum of all tree elements.

Input : Â  

Â  Â  1

/Â  Â  \

-2Â  Â  Â  3

/ \Â  Â  /Â  \

4 Â  5Â  -6 Â  2

Output : 7

Subtree with largest sum is :

-2

/ \

Also, entire tree sum is also 7.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def maxSubtreeSum(node):
    if node is None:
        return 0
    
    left_sum = maxSubtreeSum(node.left)
    right_sum = maxSubtreeSum(node.right)
    
    current_sum = node.data + left_sum + right_sum
    
    return max(current_sum, left_sum, right_sum)

# Test the algorithm with your examples

# Example 1
root1 = Node(1)
root1.left = Node(2)
root1.right = Node(3)
root1.left.left = Node(4)
root1.left.right = Node(5)
root1.right.left = Node(6)
root1.right.right = Node(7)

print(maxSubtreeSum(root1))  # Output: 28

# Example 2
root2 = Node(1)
root2.left = Node(-2)
root2.right = Node(3)
root2.left.left = Node(4)
root2.left.right = Node(5)
root2.right.left = Node(-6)
root2.right.right = Node(2)

print(maxSubtreeSum(root2))  # Output: 7
```

2.1. Construct the BST (Binary Search Tree) from its given level order traversal.

Examples:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

7

/Â  Â  \

4 Â  Â  12

/Â  \ Â  Â  /

3 Â  6Â  8

/ Â  / Â  \

1Â  5 Â  10

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructBST(level_order):
    if len(level_order) == 0:
        return None
    
    root = Node(level_order[0])
    queue = []
    queue.append(root)
    i = 1
    
    while i < len(level_order):
        current = queue.pop(0)
        
        if i < len(level_order):
            current.left = Node(level_order[i])
            queue.append(current.left)
            i += 1
        
        if i < len(level_order):
            current.right = Node(level_order[i])
            queue.append(current.right)
            i += 1
    
    return root

# Test the algorithm with your example
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(level_order)

# Perform an inorder traversal to check if the BST is constructed correctly
def inorder(node):
    if node is not None:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)

inorder(root)  # Output: 1 3 4 5 6 7 8 10 12
```

3.Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output : Yes

For the given arr[] the Binary Search Tree is:

7

/Â  Â  \

/ Â  /Â  Â  \

1 Â  5 Â  Â  10

Input : arr[] = {11, 6, 13, 5, 12, 10}

Output : No

The given arr[] do not represent the level

order traversal of a BST.

```python
def isLevelOrderBST(arr):
    n = len(arr)
    if n == 0:
        return True
    
    stack = []
    root = arr[0]
    stack.append(root)
    i = 1
    
    while i < n:
        if len(stack) == 0:
            return False
        
        if arr[i] < stack[-1]:
            stack.append(arr[i])
            i += 1
        else:
            parent = stack.pop()
            if i < n and arr[i] < parent:
                return False
            stack.append(arr[i])
            i += 1
    
    return True
```
